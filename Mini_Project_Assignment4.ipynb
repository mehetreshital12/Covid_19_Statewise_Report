```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,7))
x = df["First Dose Administered"]
  
plt.hist(x, bins = 10, color = "green")
plt.title("First Dose Administered")
plt.xlabel("First Dose Administered")
plt.ylabel("Count")
```
